```python
df13 = pd.DataFrame()
for index, coropnaam in enumerate(dfcorop["Coropnaam"]):

    df13[dfcorop.iloc[index,1]] = df.loc[coropnaam]
df13 = df13.T
## ADDING Municipality Littenseradeel, which was splitted in different municipalities in 2018
df13.loc["GM0140"] = df.loc["Noord-Friesland"].tolist()

newdict = {}
for i in df13:
    newdict[i] = "Totlabforce{}".format(i)
    
df13 = df13.rename(newdict, axis =1 )

for index, code in enumerate(df13.index.values):
    df13 = df13.rename({code: code.replace("M","")})
```

```python
df17 = pd.DataFrame()
for index, coropnaam in enumerate(dfcorop["Coropnaam"]):

    df17[dfcorop.iloc[index,1]] = df.loc[coropnaam]
df17 = df17.T
## ADDING Municipality Littenseradeel, which was splitted in different municipalities in 2018
df17.loc["GM0140"] = df.loc["Noord-Friesland"].tolist()

newdict = {}
for i in df17:
    newdict[i] = "Totlabforce{}".format(i)
    
df17 = df17.rename(newdict, axis =1 )

for index, code in enumerate(df17.index.values):
    df17 = df17.rename({code: code.replace("M","")})
    

df17 = df17.rename({"index":"Regiocode"},axis =1)
display(df17)
```

```python
df13 = pd.read_csv("Variables/Kerncijfers/beroepsbevolking-13.csv")
df13 = df13.rename({"Unnamed: 0":"Regiocode"},axis = 1)

dflabforce = df13.merge(df17,left_on = ["Regiocode"],right_index = True ,how = "outer")

dflabforce.to_csv('/Users/ruben/Documents/Datascience & Society/Thesis/Variables/Definitiveversions/Laborforce_version.csv')

dflabforce = dflabforce.set_index("Regiocode").astype(int)



for year in range(1998,2017):
    dflabforce["avglabfchange{}".format(year+1)] = (((dflabforce["Totlabforce{}".format(year+1)]) - dflabforce["Totlabforce{}".format(year)] )/dflabforce["Totlabforce{}".format(year)])*100
dflabforce = dflabforce.reset_index()
dflabforce = dflabforce.rename({"index":"Regiocode"}, axis = 1)

electionperiods = [[1999,2002,"9802"],[2002,2003,"0203"],[2003,2006,"0306"],[2006,2010,"0610"],[2010,2012,'1012'],[2012,2017,"1217"]]
dfLabforce = pd.DataFrame()


for i in electionperiods:
    startyear = i[0]+1
    avg = dflabforce[["Regiocode","avglabfchange{}".format(startyear)]]
    for j in range(i[0]+2,i[1]+1):
        avg = avg.merge(dflabforce[["avglabfchange{}".format(j),"Regiocode"]],how = "right", on = "Regiocode")
    dfLabforce["avglabfchangeelection{}".format(i[2])] = avg.iloc[:,1:].mean(axis = 1)



dfLabforce["Regiocode"] = dflabforce["Regiocode"]

dfLabforce.to_csv('/Users/ruben/Documents/Datascience & Society/Thesis/Variables/Definitiveversions/Labforce_election_version.csv')
```
